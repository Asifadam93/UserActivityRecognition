# Détection des actions de l'utilisateur avec création d'un dataset

Le but de notre projet est de réaliser un DataSet concernant les actions d'un utilisateur.

La réalisation du dataset se fera par l'intermédiaire d'une application android.

Actions à réaliser : 

* Android : 
 * Création du projet de récupération des datasets
 * Création du controller/view permettant d'enregistrer les actions de l'utilisateur (L'enregistrement se basera sur une action manuelle de l'utilisateur)
 * Sauvegarde des données sur un fichier texte
 * Ajout d'une géolocalisation
