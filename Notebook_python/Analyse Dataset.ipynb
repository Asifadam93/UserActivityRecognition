```python
# Librairies utiles
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from IPython.core.display import Image, display
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.grid_search import GridSearchCV
```

```python
# importation du dataset et affichage
activity_dataset = pd.read_csv("./dataset_fabien.csv")
activity_dataset.head(3) 
```

```python
# récupération des dummies pour les Lieux
activity_dataset = pd.get_dummies(activity_dataset, columns=['Lieux'])
activity_dataset.head(3) 

```

```python
# Liste des colonnes de notre dataset
activity_dataset.columns
```

```python
# Liste des colonnes sources et cibles
colonnes_source = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10',
       'x11', 'x12', 'x13', 'x14', 'x15', 'x16', 'x17', 'x18', 'x19', 'x20',
       'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11',
       'y12', 'y13', 'y14', 'y15', 'y16', 'y17', 'y18', 'y19', 'y20', 'z1',
       'z2', 'z3', 'z4', 'z5', 'z6', 'z7', 'z8', 'z9', 'z10', 'z11', 'z12',
       'z13', 'z14', 'z15', 'z16', 'z17', 'z18', 'z19', 'z20', 't1', 't2',
       't3', 't4', 't5', 't6', 't7', 't8', 't9', 't10', 't11', 't12', 't13',
       't14', 't15', 't16', 't17', 't18', 't19', 't20', 'Lieux_Ecole',
       'Lieux_Exterieur', 'Lieux_Maison']
colonnes_cible = ['Action']

# récupération des données correspondant aux colonnes 
source = activity_dataset[colonnes_source]
cible = activity_dataset[colonnes_cible]
```

```python
# cration d'un jeu de données de test et d'entrainement
source_train, source_test, cible_train, cible_test = train_test_split(source, cible, test_size=0.7)
```

```python
#[print('classe ',i,' : ',sum(cible_train==i)) for i in cible_train.values]

display(pd.Series.value_counts(activity_dataset['Action']))
# display(source_train, source_test, cible_train, cible_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

tuned_parameters = [{
   'hidden_layer_sizes': [(1,), (10,), (100,),(100,100),(100,100,100),(100,100,100,100),(100,100,100,100,100),(100,100,100,100,100,100),(1000,1000),(10,10)]
}]
scores = ['accuracy']

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(MLPClassifier(), tuned_parameters, cv=5, scoring=score)
    clf.fit(source_train, cible_train.values.ravel())
    
    
    print("Best parameters set found on development set:")
    print()
    print(clf.best_estimator_)
    print()
    print("Grid scores on development set:")
    print()
    for params, mean_score, scores in clf.grid_scores_:
        print("%0.3f (+/-%0.03f) for %r"
              % (mean_score, scores.std() / 2, params))
    print()

    print("Detailed classification report:")
    print()
    cible_true, cible_pred = cible_test, clf.predict(source_test)
    print(classification_report(cible_true, cible_pred))
    print()
```

```python
clf = MLPClassifier(activation='relu', alpha=0.0001, batch_size='auto', beta_1=0.9,
       beta_2=0.999, early_stopping=False, epsilon=1e-08,
       hidden_layer_sizes=(100, 100, 100, 100, 100),
       learning_rate='constant', learning_rate_init=0.001, max_iter=200,
       momentum=0.9, nesterovs_momentum=True, power_t=0.5,
       random_state=None, shuffle=True, solver='adam', tol=0.0001,
       validation_fraction=0.1, verbose=False, warm_start=False)
clf.fit(source_train, cible_train.values.ravel())
```
